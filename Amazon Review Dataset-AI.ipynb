```python
df1 = df[["name","reviews.text"]]
df1.head()
```

```python
df1["name"] = df1["name"].str.lower()
df1.head()
```

```python
df1["reviews.text"] = df1["reviews.text"].str.lower()
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df1["name"] = le.fit_transform(df1["name"])
df1.head()
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation

from nltk.stem import LancasterStemmer
lstem = LancasterStemmer()

text = df1["reviews.text"].tolist()

to_be_removed = list(punctuation) + list(stopwords.words("english"))

finaldf = []
for i in range(len(text)):
    token_word = word_tokenize(text[i])
    clean_word = [x for x in token_word if x not in to_be_removed]
    stem_word = [lstem.stem(x) for x in clean_word]
    bind_words = " ".join(stem_word)
    finaldf.append(bind_words)
    
finaldf
```

```python
finaldf = pd.DataFrame(finaldf)
finaldf.head()
```

```python
finaldf["name"]=df1["name"]
finaldf.columns = ["reviews", "name"]
finaldf.head()
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
rev = finaldf["reviews"].tolist()

for i in range(len(rev)):
    token_word = nltk.word_tokenize(rev[i])
    #print(nltk.pos_tag(token_word))
    
    adj_list = [x for x in nltk.pos_tag(token_word) if x[1] == 'JJ']
    print(adj_list)
```

```python
count_of_adj = []
counts = dict()
for j in range(len(adj_list)):
    if j in counts:
        counts[j] += 1
    else:
        counts[j] = 1
    count_of_adj.append(counts)
```
